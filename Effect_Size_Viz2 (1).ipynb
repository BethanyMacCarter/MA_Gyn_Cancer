```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
file_path='/Users/bethanymaccarter/Desktop/Effect_Size.csv'
```

```python
effect_size_df = pd.read_csv(file_path)
```

```python
effect_size_df.head()
```

```python
# Define the height of the bars
bar_height = 0.45  # Adjust this value as needed to provide more spacing between bars

# Set up the figure and axis
fig, ax = plt.subplots(figsize=(10, 10))  # Adjust the figsize as needed
effect_size_df_reversed = effect_size_df.iloc[::-1]  # Reverse the order of the DataFrame

# Define the width between the bars for "Before Matching" and "After Matching"
bar_width = 0.45  # Adjust this value as needed to provide more spacing between the bars

# Plot bars for "After Matching"
bar2 = ax.barh(np.arange(len(effect_size_df_reversed)), effect_size_df_reversed['After'], bar_height, label='After Matching', color='#1f77b4')  
# Plot bars for "Before Matching"
bar1 = ax.barh(np.arange(len(effect_size_df_reversed)) + bar_width, effect_size_df_reversed['Before'], bar_height, label='Before Matching', color='#ff7f0e')  

# Add horizontal gridlines
ax.grid(axis='x', linestyle='-', color='gray', alpha=0.5)

# Add vertical gridlines
ax.grid(axis='y', linestyle='--', color='gray', alpha=0.3)

# Set labels for the axes with bold font
ax.set_ylabel('Features', fontweight='bold')
ax.set_xlabel('Effect Size', fontweight='bold')
ax.set_title('Effect Size Before and After Matching', fontweight='bold')

# Set Y-axis limits
ax.set_ylim(-0.5, len(effect_size_df_reversed))

# Add labels, title, and legend
ax.set_yticks(np.arange(len(effect_size_df_reversed))) 
ax.set_yticklabels(effect_size_df_reversed['Feature'])  
ax.legend()

# Show plot
plt.tight_layout()
plt.show()
```
