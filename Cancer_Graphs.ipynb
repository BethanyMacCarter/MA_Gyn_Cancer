```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Data
cancer_types = ['Uterine', 'Ovarian', 'Cervical', 'Breast']
observed_ma = [201, 112, 239, 346]
observed_control = [125, 84, 303, 338]

x = np.arange(len(cancer_types))  # the label locations
width = 0.35  # the width of the bars

fig, ax = plt.subplots()
bars1 = ax.bar(x - width/2, observed_ma, width, label='Observed MA', edgecolor='black')
bars2 = ax.bar(x + width/2, observed_control, width, label='Observed Control', edgecolor='black')

# Add some text for labels, title and custom x-axis tick labels, etc.
ax.set_xlabel('Cancer Type')
ax.set_ylabel('Raw Counts (N=28,462)')
#ax.set_title('Observed Counts by Cancer Type')
ax.set_xticks(x)
ax.set_xticklabels(cancer_types)
ax.legend()

# Function to add labels on the bars
def autolabel(bars):
    for bar in bars:
        height = bar.get_height()
        ax.annotate(f'{height}',
                    xy=(bar.get_x() + bar.get_width() / 2, height),
                    xytext=(0, 2),  # 3 points vertical offset
                    textcoords="offset points",
                    ha='center', va='bottom')

autolabel(bars1)
autolabel(bars2)

# Adjust layout to ensure labels fit
fig.tight_layout()
fig.subplots_adjust(top=1.55, bottom=0.5, left=0.5, right=1.5)
#fig.subplots_adjust(top=1.05, bottom=0.4, left=0.3, right=1.1)

plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Data
cancer_types = ['Uterine', 'Ovarian', 'Cervical', 'Any Gyn', 'Breast']
risk_ratios = np.array([1.235, 1.143, 0.881, 1.047, 1.012])

# Create figure and axis
fig, ax = plt.subplots()

# Plot risk ratios as dots
ax.scatter(cancer_types, risk_ratios, color='blue')

# Add horizontal line at risk ratio = 1
ax.axhline(1, color='gray', linestyle='--')

# Add labels and title
ax.set_xlabel('Cancer Type')
ax.set_ylabel('Risk Ratio')
ax.set_title('Risk Ratios by Cancer Type')
ax.set_xticks(np.arange(len(cancer_types)))
ax.set_xticklabels(cancer_types, rotation=45, ha='right')

# Annotate points
for i, ratio in enumerate(risk_ratios):
    ax.annotate(f'{ratio:.2f}', (cancer_types[i], ratio), textcoords="offset points", xytext=(0,10), ha='center')

# Create secondary y-axis
ax2 = ax.twinx()
ax2.set_ylim(ax.get_ylim())  # Set the limits of the secondary y-axis to match the primary y-axis
ax2.set_yticks([])  # Remove ticks from the secondary y-axis

# Add vertical axis labels on the secondary y-axis
ax2.text(1.01, 0.75, 'Favors MA', transform=ax2.transAxes, ha='left', va='bottom', color='black')
ax2.text(1.01, 0.4, 'Favors Control', transform=ax2.transAxes, ha='left', va='top', color='black')

# Customize the y-axis limits
ax.set_ylim(0, max(risk_ratios) + 0.5)

# Show plot
plt.tight_layout()
plt.show()
```
